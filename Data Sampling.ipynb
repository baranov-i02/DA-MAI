```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import scipy.io
A=scipy.io.loadmat('Level1_Data2.mat')
D=A['D']
print(D)
N= 100
Dmin=np.min(D)
Dmax=np.max(D)
Dmean=np.mean(0)
Hpoint=np.linspace(Dmin,Dmax,N+1)
Count=[]

plt.figure(figsize=(17,8))
for i in range(1,N+1):
    CountNow=np.count_nonzero(D<Hpoint[i])-np.count_nonzero(D<Hpoint[ i-1])
    Count.append(CountNow)
    plt.fill([Hpoint[i-1],Hpoint[i-1],Hpoint[i] ,Hpoint [i], Hpoint[i-1]],[0,CountNow, CountNow, 0 , 0])
plt.xlabel('x')
plt.ylabel ('Количество')
plt.xlim([Dmin-0.2*(Dmax-Dmin) ,Dmax+0.2* (Dmax-Dmin)])
plt.title('Гистограмма')
plt.grid()
plt.legend(["f(x)"])
plt.show()

print('Распределение нормальное')
M = np.sum(D)/np.shape(D)[1]
print('Мат.ожидания: {0}'.format(M))
Disp = np.sum([(Xi - M)**2 for Xi in D])/np.shape(D)[1]
print('Дисперсия: {0}'. format(Disp))
print('Сигма: {0}'.format (Disp**0.5))
```

```python
A=scipy.io.loadmat('LineMNK_Data2.mat')
X=np.array(A['X'])
Y=A['Y']

Dmin=np.min(X)
Dmax=np.max(X)
Sx= X.sum()
Sy= Y.sum()
X2=X*X
XY=X*Y
SX2=X2.sum()
SXY=XY.sum()

K,B=sp.symbols('K,B')
f1=sp.Eq(2*K*Sx+2*B*len(X)-2*Sy,0)
f2=sp.Eq(2*K*SX2+2*B*Sx-2*SXY,0)
KB=sp.solve([f1,f2],[K,B])

print(KB)

YClear=[KB[K]*x+KB[B] for x in X]
YClear = np.array(YClear)

plt.figure(figsize=(17,8))
for x,y in zip(X,Y):
    plt.scatter(x,y)
plt.plot(X[0][:-1],YClear[0][:-1],color = 'red')
plt.xlabel('x')
plt.ylabel('Количество')
plt.xlim([Dmin-0.2*(Dmax-Dmin),Dmax+0.2*(Dmax-Dmin)])
plt.title('Гистограмма')
plt.grid()
plt.legend(["f(x)"])
```
